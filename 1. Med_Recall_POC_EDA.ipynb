```python
import pandas as pd
import json
```

```python
with open("./Data/device-pma-0001-of-0001.json", "r") as read_file_1:
    data = json.load(read_file_1)
df_pma = pd.DataFrame(data['results'])
```

```python
with open("./Data/device-recall-0001-of-0001.json", "r") as read_file_2:
    recall = json.load(read_file_2)
df_recall = pd.DataFrame(recall['results'])
```

```python
# add the metadata to the dataset 
df_pma=df_pma.join(df_pma['openfda'].apply(pd.Series))
```

```python
df_pma.dropna(subset=['fei_number'],inplace=True)
```

```python
df_pma=df_pma.reset_index(drop=True)
```

```python
# obtain a list product recall fei > 5 to remove bad values
result = [item for item in df_recall['firm_fei_number'].tolist() if len(item)>5]
```

```python
df_pma['recall_status']=0
indx=df_pma.index
count=-1
for xs in df_pma['fei_number']:#.head(500):
    count=count+1
    
    for x in xs:
         if x in (result):
           # print(x)
            df_pma.set_value( indx[count],'recall_status',1)
            break
```
